### 중요  : AND 학습하기  아까 같은 문제를 텐서플로로 풀어보자

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

* Hyperparameters

```python
tf.set_random_seed(777)  # for reproducibility 랜덤이 나오게는 하는데 그 랜덤의 seed를 고정해버리면 어느컴퓨터에서든 랜덤 돌려도 똑같은 숫자 나옴.
learning_rate = 0.1
```

* dataset

```python
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]

y_data = [[0],
          [0],
          [0],
          [1]]
```

```python
# x_data.shape 넘파이실행 전에 하라고 하면 오류뜸
```

```python
# and gate 학습하는 코드를 가지고 XOR 을 돌려보자  선하나 긋는 애 가지고 xor돌려보니 결과로 가보면 0.5가 나옴.. 학습이 안된거 원래는 0이나 1로 가깝게 나와야 하는데. cost도 안줄어들고 결국 사람인 내가 잘못 지정해준거임. 레이어하나만 가지고는 xor데이터는 못한다.
'''
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]

y_data = [[0],
          [1],
          [1],
          [0]]
          
'''
```

```python
x_data = np.array(x_data, dtype=np.float32)   #파이썬으로 만든 리스트를 넘파이(수학라이브러리)한테 플로트 타입으로 해서 데이터를 써보자고 하는 것.
y_data = np.array(y_data, dtype=np.float32)
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
X = tf.placeholder(tf.float32, [None, 2])  # 텐서플로 쓰려는 약속. placeholder는 아직 x가 안들어왔는데 자리를 맡아놓는거. 텐서플로에서 미리 x를 줄 수는 없음 수천장의 고양, 개 사진 있어 이거 한번에 부으면 컴이 감당못함 그래서 None은 나중에 넣어준다는 말. 2개의 성분이고.
Y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2, 1]), name='weight1')  #웨이트가 2개
b1 = tf.Variable(tf.random_normal([1]), name='bias1')  #바이어스는 1개                     이런 레이어 하나로는 xor학습 못함
hypothesis = tf.sigmoid(tf.matmul(X, W1) + b1)
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
# Launch graph
sess = tf.Session()
```

```python
# TensorFlow 변수들(variables) 초기화 (Initialization)
sess.run(tf.global_variables_initializer())
```

```python
for i in range(10001):
    sess.run(train, feed_dict={X: x_data, Y: y_data})

    if i % 1000 ==0:
        c1 = sess.run(cost, feed_dict={X: x_data, Y: y_data})
        print('step={} / cost={}'.format(i, c1))
```
