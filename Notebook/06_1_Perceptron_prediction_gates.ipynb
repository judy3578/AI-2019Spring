### Perceptrons - Making Predictions
#### Creating a gate with Perceptron

```python
import numpy as np # 파이썬을 가지고 수학적으로 계산을 하기 쉽게 해주는 넘파이 파이썬으로 수학을 안쓰는 경우도 있기 때문에 이렇게 갔다쓰게 함 np는 망치같이 나중에 계산할때 편하게 해주는 툴
```

### AND Gate

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.1, 0.9])  # 원래는 0.5, 0.5  이번에는 w를 0.1, 0.9로 바꿔보자   그리고 둘이 합쳐서 1이 되어야함 왜냐하면 어차피 나중에 1을 만점을 볼거기 때문에 3을 넣어도 나눠서 1을 만점으로 계산함.
    b = -0.7  # -0.7 을 -0.5 로 바꾸고 돌려보기  그랬더니 값이 애매해짐  -2 로 바꿔보고 또 돌려보자
    # activation 계산 이 아래값이 activation 값이 됨! 이게 0보다 작으면 0이고 크면 1 아직 x1 x2는 없음 입력받아야지~
    tmp = w[0]*x[0] + w[1]*x[1] + b # tmp = np.sum(w*x) + b
    print(tmp)
    
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND(1,0) # x1 x2를 다르게 넣어서 결과를 봐보자 
```

```python
AND(1,1)  # -2 로 bais 를 주고 돌려봤더니 라면 김밥을 둘다 줘도 이미 너무 화나있어서 기분이 좋아지지 않음. 
```

```python
AND(0,0)   # 둘다 0이 되니 김밥도 라면도 없으니 더 작음 AND(1,0)보다
```

```python
AND(0,1)   # w를 0.1, 0.9로 바꾸면 라면만 있으면 그래도 기분이 좋음. 이때도 이미 기분이 너무 나쁘면 돌아오지는 않지만
```

### NAND Gate

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = w[0]*x[0] + w[1]*x[1] + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(1,0)
```

```python
NAND(1,1)
```

### OR Gate

```python
# b만 바꿨는데 왜 or gate가 됐을까? --> 시험
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(1,0)
```

```python
OR(0,1)
```

```python
OR(1,1)
```

### XOR Gate  동일한데 

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,1)
```

```python
XOR(0,1)
```

```python
XOR(1,0)
```

XOR cannot be expressed as a single layer Perceptron.
