### Perceptrons - Making Predictions

Assuming that we have  이게 있다고 가정하자 이게 왜 나온건지는 다음에 그냥 숫자 가져왔다고 하자
```
weights = [-0.1, 0.206, -0.234]     이걸 주어졌는데 이게 어디서 온걸까? 전에 숫자찍어보는 게임 개념을 가지고 w를 구하는걸 알아보자  06_03
```

````
def predict2(X, w):   # 입력과 노드가 들어옴  w[0]은 bias
    activation = w[0] + w[1]* X[0] + w[2]* X[1]  # w는 0,1,2 있는걸 보니 3개가 있는데 X는 0,1 2개 밖에 없음 즉 쌍이 안맞는 거임 맨앞에 애는 X없이 그냥 w만 옴. 입력으로 들어온 노드는 X0과 X1 밖에 없음 weight는 3개가 들어오는데 X가 안붙는건 그냥 bias임. 같이
    # 처리하려고 w에 넣은거임. 다 더한거는 activation이라고 부르기로 했고 
    if activation >= 0.0:  #다 더한게 0보다 크면 1이라고 해주고 아니면 0
        return 1.0
    else:
        return 0.0
      
      # bias = w[0]
````

````
# test predictions    그리고 데이터를 하나 만들었음. 인터넷에서 가져옴. 앞에 2.78 , 2.55 = X1 X2 는 0이라고 이렇게 답을 정해주고 0이면 김밥 1이면 라면이다.매 점들마다 0인지 1인지 컴이 확인할거임/  10개의 데이터와 정답을 알려줌
dataset = [[2.7810836,2.550537003,0],
    [1.465489372,2.362125076,0],
    [3.396561688,4.400293529,0],
    [1.38807019,1.850220317,0],
    [3.06407232,3.005305973,0],
    [7.627531214,2.759262235,1],
    [5.332441248,2.088626775,1],
    [6.922596716,1.77106367,1],
    [8.675418651,-0.242068655,1],
    [7.673756466,3.508563011,1]]
````

````
bias = -0.1
w0 = 0.206
w1 = -0.234
weights = [bias, w0, w1]   #[-0.1, 0.206, -0.234]
print(weights)
````

````
for row in dataset:   #데이터를 한줄씩 출력하는 문법 파이썬 문법익숙해지기 위해 쓴 코드
    print(row[0], row[1], row[2], row[-1])  # print(row)
````

````
for row in dataset:
    prediction = predict2(row, weights) # predict2에 row와 weihts를 넣음 그리고 그 값을 프레딕션 변수에 저장. 그리고 row[-1]은 맨 마지막거를 출력함. 데이터가 백개가 되면 끝에꺼 출력해서 이렇게 써주고
    answer = row[-1]
    print("Expected={}, Predicted={}".format(answer, prediction))  # 다 맞췄음!  눈으로 쉽게 확인하기 위해 그래프를 그려보자 스스로 할줄 알기
````

````
for j in range(10):  #(5)
  for i in range(99):
    x = i * 0.1 # 0.1씩 테스트 하고 싶어서 0.1 곱함
    y = j   # 으로 고정함
    prediction = predict2([x, y], weights)

    if prediction == 0:
      plt.plot(x, y, 'b.') # 이렇게 하면 점들이 찍힘
    else : 
      plt.plot(x, y, 'r.')
````
