## Basic data types

```python
x = 3
print(type(x)) # Prints "<class 'int'>"
print(x)       # Prints "3"
print(x + 1)   # Addition; prints "4"
print(x - 1)   # Subtraction; prints "2"
print(x * 2)   # Multiplication; prints "6"
print(x ** 2)  # Exponentiation; prints "9"
x += 1
print(x)  # Prints "4"
x *= 2
print(x)  # Prints "8"
```

```python
y = 2.5
print(type(y)) # Prints "<class 'float'>"
print(y, y + 1, y * 2, y ** 2) # Prints "2.5 3.5 5.0 6.25"
```

```python
x = 1
print(x)
print(x+1)
x += 1
print(x)
print(type(x))
print(type(2.555))
print(type("s"))
2 ** 3

```

## Operators

### + operator

```python
print((1, 2, 3) + (4, 5, 6))
print([1, 2, 3] + [4, 5, 6])
print("Hello" + " " + "World")
```

### * operator 별모양 표시
The * operator produces a new tuple, list, or string that "repeats" the original content.

```python
print((1, 2, 3) * 3)
print([1, 2, 3] * 3)
print("Hello "*3)
```

## Containers
Python includes several built-in container types: lists, dictionaries, sets, and tuples.

### Tuple
A simple immutable (변경할 수 없는, 불변의) ordered sequence of items

```python
# creating a tuple
months = ('January','February','March','April','May','June',\
'July','August','September','October','November','December')

print(months[0])
print("index of 7 ==> " , months[7])
```

To print one by one

```python
# iterate through them:
for item in months:
    print (item)
```

```python
for i in range(12):
  print(months[i])
```

```python
t = ('john', 32, (2,3,4,5), 'hello') # 튜플속에 튜플도 가능
print(t)
print(t[2])
print(t[2][1])
print("t[:2]= " , t[:2]) # index NOT included 데이터 불러올때 이런 약속 쓸 수 있음 시험! 인덱스 찾아가는 거
print(t[2:]) # index included O

print(t[-1])
print(t[-2])
print(t[-4])
print(t[3])


# t[2] = 1 튜플은 고칠 수가 없음
```

### List
Mutable (바꿀수 있는, 변경가능한) ordered sequence of items of mixed types

```python
li = ['hallym', 1, 3.141572, 'hello']
print(li)
li[1] = 45
print(li)
li.append('September')
print(li)
li.append('good')
print(li)
```

```python
li2 = ['a', 1, 2, ['b', 'c']]
print(li2)
print(li2[3][1])
```
